## Joining the postcodes with coordinates

First, I read in the previous dataframe

```python
import pandas as pd
df_grouped = pd.read_csv('Toronto_PostCodes.csv', index_col = 0)
df_grouped.head()
```

I tried using **geocoder** but it took forever to even get a single postcode (see below), so I am reading in the CSV.

```python
df_coord = pd.read_csv('http://cocl.us/Geospatial_data')
df_coord.head()
```

## Merging the two dataframes to get a df with Postcodes, Names, and Coordinates

```python
df_merged = pd.merge(df_grouped, df_coord, how='left', left_on='Postcode',right_on='Postal Code')
df_merged
```

```python
df_merged.to_csv('Toronto_Merged.csv')
```
